```python
import math
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import MDS

from sklearn_extra.cluster import KMedoids
from sklearn.cluster import KMeans

from mpl_toolkits.mplot3d import Axes3D

```

```python
stockData = pd.read_csv('stockData.csv')

stockData.head()
```

**Loading Data**

**NP array of Size 165 X 4277 (165 Stocks, 4277 Days**

```python
sNp = np.array(stockData.iloc[:, 1:])

stockNpArray = np.transpose(sNp)

np.shape(stockNpArray)
```

**Step 1: Getting Log Returns**

```python
def getLogDiffAllStocks(matrix):
  logDiffMatrix = []

  for row in matrix:
    logDiffMatrix.append(np.diff(np.log(row)))
  
  return logDiffMatrix
```

```python
#After Log Differencing each stock

logDiffMatrix = getLogDiffAllStocks(stockNpArray)

np.shape(logDiffMatrix)
```

**Step 2: Dividing into Short time-intervals(Epochs)**

426 Epochs Each of Size 165 X 20

```python
def getEpochs(matrix, size = 20, slide = 10):
  finalEpochs = []

  startIdx = 0

  while startIdx+size < len(matrix[1]):
    epoch = []

    for row in matrix:
      epoch.append(row[startIdx: startIdx+size])
    
    finalEpochs.append(epoch)

    startIdx+= slide
  
  return finalEpochs
```

```python
epochList = getEpochs(logDiffMatrix)

print(np.shape(epochList)) 

print(np.shape(epochList[0]))
```

**Step 3: Get Correlation Matrices After Noise Suppression**

Finally, we get 426 correlation matrices of size 165 X 165 each.

```python
def noiseSuppressSingleEl(el, epsilon):
  sign = -1 if el < 0 else 1

  return sign* math.pow(abs(el), 1+ epsilon)

# noiseSuppressSingleEl(-3, 0.6)
```

```python
def noiseSuppressMatrix(matrix, epsilon):
  newMatrix = np.zeros(shape = (len(matrix), len(matrix)))

  for i in range(len(matrix)):
    for j in range(len(matrix)):
      newMatrix[i][j] = noiseSuppressSingleEl(matrix[i][j], epsilon)
    
  return newMatrix
```

```python
def getAllCorrelationMatrices(epochList, epsilon):
  corrMatrices = []

  for epoch in epochList:
    correlMatrix = np.corrcoef(epoch)
    noiseSuppressedCorrMatrix = noiseSuppressMatrix(correlMatrix, epsilon)
    corrMatrices.append(noiseSuppressedCorrMatrix)
  
  return corrMatrices
```

```python
corrMatrices = getAllCorrelationMatrices(epochList, 0.6)

np.shape(corrMatrices)
```


**Step 4: Build Similarity Matrix**

We get Similarity Matrix of Size 430 X 430


```python
def getSimilarityMatrix(corrMatrices):
  similarityMatrix = np.zeros(shape = (len(corrMatrices), len(corrMatrices)))

  for i in range(len(corrMatrices)):
    for j in range(len(corrMatrices)):
      similarityMatrix[i][j] = np.mean(np.abs(corrMatrices[i] - corrMatrices[j]))
  
  return similarityMatrix
```

```python
similarityMatrix = getSimilarityMatrix(corrMatrices)
```

```python
np.shape(similarityMatrix)
```

```python
def getDistanceMatrix(matrix):
  distanceMatrix = np.zeros(shape = (len(matrix), len(matrix)))
  for i in range(len(matrix)):
    for j in range(len(matrix)):
      distanceMatrix[i][j] = math.sqrt(2* (1- matrix[i][j]))
  return distanceMatrix
```

```python
distanceMatrix = getDistanceMatrix(similarityMatrix)
```

# K-Means Clustering

```python
meanCorrelVector = []

for matrix in corrMatrices:
    meanCorrelVector.append(np.mean(matrix))
    
meanCorrelVector = np.array(meanCorrelVector)
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(meanCorrelVector.reshape(-1,1))
```

```python
labels = kmeans.labels_
```

```python
print((labels))
```

```python
adjustedLabels = []

for l in labels:
    adjustedLabels.append(l+1)

print(adjustedLabels)
```

# Transition Probabilities

```python
transitionProbMatrix = np.zeros(shape = (4, 4))

for i in range(1, len(labels)):
    currLabel = labels[i]
    prevLabel = labels[i-1]
    
    transitionProbMatrix[prevLabel][currLabel]+=1

print(transitionProbMatrix)
```

```python
rowSums = []
for row in transitionProbMatrix:
    rowSums.append(np.sum(row))

print(rowSums)
```

```python
for i in range(4):
    for j in range(4):
        transitionProbMatrix[i][j]/= rowSums[i]
```

```python
print(np.matrix(transitionProbMatrix))
```
